Indexing 2d array (Matrix)

```python
arr_2d = np.array([[5,10,15], [20,25,30], [35,40,45]])
arr_2d
```

```python
arr_2d[0][0]
```

```python
arr_2d[0,0]
# better than double brackets
```

```python
arr_2d[:2,1:]
# all rows to 2
# all columns from 1
```

Conditional selection

```python
arr = np.arange(1, 11)
arr
```

```python
arr > 5
# returns boolean array
```

```python
bool_arr = arr > 5
```

```python
arr[bool_arr]
```

```python
arr[arr > 5]
#same
```

```python
arr_python = [1,2,3,4,5]
arr_python
```
