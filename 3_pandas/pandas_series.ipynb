```python
import pandas as pd
import numpy as np
```

```python
labels = ['a', 'b', 'c']
my_data = [10, 20, 30]
arr = np.array(my_data)
d = {'a':10, 'b':20, 'c':30}
```

```python
arr
```

```python

pd.Series(data=arr)
# basically an indexed np array
```

```python
pd.Series(data=arr, index=labels)
```

Dict to Series

```python
pd.Series(d)
# takes key as label
```

```python
pd.Series(data=[sum, len, print])
# Series can pretty much hold every type even function
```
