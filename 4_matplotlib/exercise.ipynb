```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(0, 100)
y = x*2
z = x**2
```

- Create a figure object
- Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax
- Plot(x, y) on that axes and set the labels and titles

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("title")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x, y)
ax2.plot(x, y)
```

Create the plot below by adding two axes to a figure object at [0,0,1,1] and [.2,.5,.4,.4]

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

ax1.plot(x, z, color="green")
ax1.plot(x, y)
ax1.set_xlabel("x")
ax1.set_ylabel("z")

ax2.plot(x, z)
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("zoom")
ax2.set_xlim([20, 22])
ax2.set_ylim([300, 500])
```

Use plt.subplots(nrows=1, ncols=2) to create the plot below

```python
fig, axe = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axe = plt.subplots(nrows=1, ncols=2)
axe[0].plot(x, y,  linestyle='--')
axe[1].plot(x, z, color='red', linewidth=5)
```

See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.

```python
fig, axe = plt.subplots(nrows=1, ncols=2, figsize=[12, 3])
axe[0].plot(x, y,  linewidth=5)
axe[1].plot(x, z, color='red', linestyle='--')
```
