```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x,y
```

Functional method of creating plots

```python
plt.plot(x, y)
plt.show() # if not in jupyter notebook
```

```python
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.plot(x, y)
```

Multi plots on same canvas

```python
plt.subplot(1, 2, 1) # rows, columns, plotnumber
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(y,x,'b')
```

Object oriented (better)

```python
fig = plt.figure() # figure object

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # list with -> left, bottom, width, heigt of the axis

axes.plot(x, y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure() # basically an empty canvas

# then we add axis to this canvas
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

# Part 2 - Subplot with OO

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # rows, cols

#axes.plot(x,y)
```

```python
axes
# array of axes
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) # rows, cols
for current_ax in axes:
    current_ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2) 
axes[0].plot(x,y)
axes[0].set_title("First Plot")
# indexing works too
```

```python
plt.tight_layout() # will avaoid overlapping
```

Fig size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# Part 3 - Customizing

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='red')
```

Line width

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='red', linewidth=20)
```

Transparency

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='red', linewidth=20, alpha=0.5)
```

Line style

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='red', linewidth=20, alpha=0.5, linestyle='--')
```

Markers

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='red', linewidth=20, alpha=0.5, linestyle='--', marker='o', markersize=10, markerfacecolor='black', markeredgecolor='yellow', markeredgewidth=20)
```

Plot range

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, color='red', linewidth=2, alpha=0.5, linestyle='--')

ax.set_xlim([0, 1]) # x 
ax.set_ylim([0, 2]) # y
```
