```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x,y
```

Functional method of creating plots

```python
plt.plot(x, y)
plt.show() # if not in jupyter notebook
```

```python
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.plot(x, y)
```

Multi plots on same canvas

```python
plt.subplot(1, 2, 1) # rows, columns, plotnumber
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(y,x,'b')
```

Object oriented (better)

```python
fig = plt.figure() # figure object

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # list with -> left, bottom, width, heigt of the axis

axes.plot(x, y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
```

```python
fig = plt.figure() # basically an empty canvas

# then we add axis to this canvas
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```
