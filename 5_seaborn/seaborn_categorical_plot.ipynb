```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Barplot - group by action visualization with mean

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
# average total bill for sex
# estimator wants an estimator functino, like np.std
```

Countplot - bar plot with counting number of occurences

```python
sns.countplot(x='sex', data=tips)
```

Boxplot - Show distribution of categorical data

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
# people on friday tend to have a bigger bill when they non smokers
```

Violinplot - similar to boxplot. Distribution per category

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

Stripplot - Scatterplot where 1 variable is categorical

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

Swarmplot = Violinplot + Stripplot

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

Factorplot - Most general form of all these plots

```python
sns.factorplot(x='day', y='total_bill', data=tips ,kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips ,kind='violin')
```
