```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

Dist plot (Distrbution for 1 variable)

```python
sns.displot(tips['total_bill'], kde=True, bins=30)
```

Join plot (combine two distplots)

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg') # col, col, dataframe
```

Pairplot (pairwise relationship for entire dataframe) - does joinplot for every col combination

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
# hue = categorical column
```

Rugplot - is a histogram without count, draws an dashmark for every single point

```python
sns.rugplot(tips['total_bill'])
```

KDE plot - distplot + rugplot

```python
sns.kdeplot(x='total_bill', y='tip', data=tips)
```
