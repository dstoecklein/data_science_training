```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

Heatmaps - Data should always be in matrix form

```python
tips.corr()
# now the column and index are the same, needed for heatmap
```

```python
tc = tips.corr()
```

```python
sns.heatmap(data=tc, annot=True, cmap='coolwarm')
```

```python
flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(data=fp, linewidths=3)
```

Clutermap - hierarchical clustering

```python
sns.clustermap(data=fp, cmap='coolwarm', standard_scale=1)
```
