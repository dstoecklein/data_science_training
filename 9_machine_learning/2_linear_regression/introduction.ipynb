# Linear Regression
<br>
Example: Shaq is a tall basketball player. Chances that his son is also tall is high. In reality he is only nearly as tall.
<br>
This is called regression, as in "A fathers sons height tends to regress the mean (average) height"

In the simplest possible Example with 2 datapoints:
- All we are trying to do when calculating the regression line, is draw a line that is as close to every dot as possible

![](linear_reg_2_dots.jpg)

Wouldnt it be cool if we could apply this same method to a graph with more than just two datapoints?
- By doing this, we could take multiple men and their son's heights and tell a man how tall we expect his son to be... before he even has a son
- The goal is to minimize the vertical distance between all data points

![](linear_reg.jpg)

Can use Least Squares Method to minimize the vertical dinstance
