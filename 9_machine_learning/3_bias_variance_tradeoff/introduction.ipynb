Bias Variance Trade Off is a fundamental topic of understanding your models performance <br>
- It is the point where we are adding just noise by adding model complexity
- The training error goes down as it has to, but the test error is starting to go up
- Model begins to overfit

Imagine that the center of the target is a model that perfectly predicts the correct values <br>
As we move away from the bulls-eye, the predictions get worse and worse

![](bullseye.jpg)

We aiming for low variance and low bias <br>
In reality we trade of variance for bias
