```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv("classified_data", index_col=0)
```

```python
df.head()
```

Anonymised data - we dont know what those values and cols are

The scale in KNN matters

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = df.drop("TARGET CLASS", axis=1)
y = df["TARGET CLASS"]
```

```python
scaler.fit(X=X)
```

```python
scaled_features = scaler.transform(X=X)
```

```python
scaled_features
```

```python
df_feat = pd.DataFrame(data=scaled_features, columns=X.columns)
```

```python
df_feat.head()
# scaled
```

```python
X.head()
# old
```

# Model training

```python
X_train, X_test, y_train, y_test = train_test_split(df_feat, y, test_size=0.3, random_state=101)
```

```python
model = KNeighborsClassifier(n_neighbors=1)
```

```python
model.fit(X=X_train, y=y_train)
```

```python
predictions = model.predict(X=X_test)
```

```python
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

# Make it better by finding better K using Elbow Method

```python
error_rate = []
```

```python
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X=X_train, y=y_train)
    pred_i = knn.predict(X=X_test)
    error_rate.append(np.mean(pred_i != y_test)) # average where preds are not equal to actual test value
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40), error_rate, color="blue", linestyle="dashed", 
    marker="o", markerfacecolor="red", markersize=10)
plt.title("Error Rate vs K value")
plt.xlabel("K")
plt.ylabel("Erro Rate")
```

```python
model = KNeighborsClassifier(n_neighbors=17)
model.fit(X=X_train, y=y_train)
predictions = model.predict(X=X_test)
print(confusion_matrix(y_test, predictions))
print()
print(classification_report(y_test, predictions))
```
