## Classification algorithm <br>
![](knn_1.jpg)

- calculate distance from x to all points in your data
- sort points in your data by increasing distance from x
- predict the majority label of the "k" closest point

![](knn_2.jpg)

k = 3, k = 6

So, choosing a K will affect what class a new point is assigned to 

Pros:
- Very simple
- Training is trivial
- Few parameters ("K", "Distance Metric")
- Works with any number of classes
  
Cons:
- High prediction cost (worse for large data sets)
- not good with high dimensional data
- categorical features wont work
