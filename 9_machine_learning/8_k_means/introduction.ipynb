Unsupervised learning algorithms
- Attempt to group similar clusters together in the data

Goal:
- divide data into distinct groups such that samples within each group are similar

How it works:
- Choose a number of cluster "K"
- randomly assigned each sample to a cluster
- repeat:
- - For each cluster compute the cluster centroid by taking the mean vector of points in the cluster
- Assign each sample to the cluster for whic the centroid is the closest

![](kmean_1.jpg)

Choosing a K value:
- isnt simple
- Using elbow method
- - Compute sum of squared error (SSE) for some values K (e.g. 2,4,6, ...)
- - Choose the K at which SSE decrease abruptly
- - By increasing the number of cluster, we wont get any more information
