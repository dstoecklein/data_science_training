```python
X = df.drop("benign_0__mal_1", axis=1)
y = df["benign_0__mal_1"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train) # fit + transform
```

```python
X_test = scaler.transform(X=X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(30, activation="relu"))
model.add(Dense(15, activation="relu"))
model.add(Dense(1, activation="sigmoid")) # because this is a binary classification problem
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
model.fit(x=X_train, y=y_train, epochs=600, validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

Overfitting <br>
- train loss constantly goes down
- however, val_loss goes up
- We trained to much

# Early stopping

```python
model = Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dense(15, activation="relu"))
model.add(Dense(1, activation="sigmoid")) # because this is a binary classification problem
model.compile(loss="binary_crossentropy", optimizer="adam")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
# monitor the validation loss and we try to minimize the thing we monitoring
# we wait 25 epochs because of noise that could occur
# Accuracy = maximize
# Loss = minimize
```

```python
model.fit(
    x=X_train, 
    y=y_train, 
    epochs=600, 
    validation_data=(X_test, y_test), 
    callbacks=[early_stop]
)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

Early stop was triggered

# Dropout

```python
model = Sequential()
model.add(Dense(30, activation="relu"))
model.add(Dropout(rate=0.5)) # 50% of neurons each batch are going to be randomly turned off
model.add(Dense(15, activation="relu"))
model.add(Dropout(rate=0.5))
model.add(Dense(1, activation="sigmoid")) # because this is a binary classification problem
model.compile(loss="binary_crossentropy", optimizer="adam")

early_stop = EarlyStopping(monitor="val_loss", mode="min", verbose=1, patience=25)
```

```python
model.fit(
    x=X_train, 
    y=y_train, 
    epochs=600, 
    validation_data=(X_test, y_test), 
    callbacks=[early_stop]
)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

Thats what we want to see

# Full evaluation

```python
predictions = model.predict(X_test)
# https://stackoverflow.com/questions/68776790/model-predict-classes-is-deprecated-what-to-use-instead
predictions = np.where(predictions > 0.5, 1, 0)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
