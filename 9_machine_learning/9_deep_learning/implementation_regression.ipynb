# ML part

```python
X = df.drop(["date", "price"], axis=1).values
y = df["price"].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train) # fit + transform
```

```python
X_test = scaler.transform(X_test) # only transform
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
model.add(Dense(19, activation="relu"))
model.add(Dense(19, activation="relu"))
model.add(Dense(19, activation="relu"))
model.add(Dense(19, activation="relu"))
model.add(Dense(1))
```

```python
model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(
    x=X_train, 
    y=y_train, 
    epochs=400, 
    validation_data=(X_test, y_test), 
    batch_size=128
) 
# we train on train data, but as we go along, we test after each epoch on test data
```

# Model evaluation

now we also have "val_loss" in the history

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
mean_squared_error(y_test, predictions) # large value cause we squaring the house price
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
mean_absolute_error(y_test, predictions) # on average we are off by 11.000 dollars
```
