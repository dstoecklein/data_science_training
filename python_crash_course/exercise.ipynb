Create a function that returns the domain of a given email

```python
def grab_domain(email: str):
    domain = email.split("@")
    return str(domain[1])
print(grab_domain("user@domain.com"))
```

Create a function that returns True if the word 'dog' is contained in the input string. Account for capitalization.

```python
def findDog(string_: str):
    return True if 'dog' in string_.lower() else False
findDog('Is there a dog here?')
```

Create a function that counts the number of times the word 'dog' occurs in a string.

```python
def countDog(string_: str):
    arr_ = string_.lower().split(" ")
    return arr_.count('dog')
countDog('This dog runs faster than the other dog dude!')
```

You are driving a little too fast and a police officer stops you. Write a function to return one of 3 possible results: "No Ticket", "Small Ticket", or "Big Ticket". If your speed is 60 or less, the result is "No ticket". If speed is between 61 and 80 inclusive, the result is "Small ticket". Is speed is 81 or more, the result is "Big ticket". Unless it is your birthday (encoded as a boolean value in the parameters of the function) -- on your birthday, your speed can be 5 high in all cases.

```python
def caught_speeding(speed, is_birthday):
    b_speed = 0
    if is_birthday:
        b_speed += 5
    
    if speed <= (60 + b_speed):
        return "No ticket"
    elif (61 + b_speed) <= speed <= (80 + b_speed):
        return "Small ticket"
    elif speed > (80 + b_speed):
        return "Big ticket"
caught_speeding(81, False)
```
